 DF_MUN_2021 - MEDIDAS DE POSIÇÃO - DISPERSÃO DATAFRAME COM AS MÉDIAS DOS MUNICÍPIOS

```python
#---------------------------------
# DATAFRAME df_mun_2021  92 x 2 
#---------------------------------
# MEDIDAS DE POSIÇÃO / DISPERSÃO
#---------------------------------
soma= df.sum()
print(soma)
soma = df_mun_2021['taxa'].sum
media= df_mun_2021['taxa'].mean()
min =  df_mun_2021['taxa'].min()
max =  df_mun_2021['taxa'].max()
mediana= df_mun_2021['taxa'].median()
moda1= df_mun_2021['taxa'].mode()         # Função MODA no PANDAS
moda1.shape                               # CADA MUNIÇIPIO TEM UMA MODA (92)
n=len(moda1)
print(n)                
#moda2 = stats.mode(df_mun_2021['taxa'])# Função MODA no SCIPY
var= df_mun_2021['taxa'].var()          # VARIÂNCIA AMOSTRAL (PADRÃO)
dp = df_mun_2021['taxa'].std()          # DESVIO PADRÃO AMOSTRAL
cv=dp/media                             # CV = DP/MÉDIA (AMOSTRAL) 
```

```python
#--------------------------------------------------------
# MEDIDAS DE POSIÇÃO / DISPERSÃO - IMPRESSÃO - ANO 2021
#--------------------------------------------------------
print()
print("soma:{}".format(soma))
print()
print("MÉDIA,.....valores de 2021")
print("média:{:.4f}".format(media))
print("min:{:.4f}".format(min))
print("max:{:.4f}".format(max))
print("mediana:{:.4f}".format(mediana))
print("moda1:{:.4f}".format(moda1[0]))
#print("moda2:{:.4f}".format(moda2)) 
print()
print("VARIÂNCIA-DESVIO PADRÃO-CV (AMOSTRAIS)")
print("Varamt:{:.4f}".format(var))
print("DPamt:{:.4f}".format(dp))
print("CVamt:{:.4f}".format(dp/media))   # COEFICIENTE DE VARIAÇÀO = DESVIO PADRÃO/MÉDIA 
```

### > VARIÂNCIA DA POPULAÇÃO (2021) - PACOTE **STATISTICS**

```python
import statistics as st
import math as mat
varp = st.pvariance(df_mun_2021['taxa'])  # VARIÂNCIA POPULAÇÃO 
print("varpop:{:.4f}".format(varp))  
print("varamt:{:.4f}".format(var))  
print("var/varp:{:.15f}".format(var/varp)) # RELAÇÃO ENTRE VAR/VARP = n/n-1 
print("92/91:   {:.15f}".format(92/91))    # FATOR DE CORREÇÃO DE BESSEL ((n/(n-1)))
dpp = st.pstdev(df_mun_2021['taxa'])       # DESVIO PADRÃO DA POPULAÇÃO 
print("dppop:{:.4f}".format(dpp))           
print("dppop:{:.4f}".format(mat.sqrt(varp)))
print("CV_POP:{:.4F}".format(dpp/media))          
```

###  QUARTIS - ASSIMETRIA / CURTOSE - DF_MUN_2021

```python
#----------------------------------
# QUARTIS - ASSIMETRIA / CURTOSE
#----------------------------------
q1  = df_mun_2021['taxa'].quantile(q=0.25)
q2  = df_mun_2021['taxa'].quantile(q=0.50)
q3  = df_mun_2021['taxa'].quantile(q=0.75)
p10 = df_mun_2021['taxa'].quantile(q=0.10)
p20 = df_mun_2021['taxa'].quantile(q=0.20)
p90 = df_mun_2021['taxa'].quantile(q=0.90)

print("Q1:{:.4f}".format(q1))
print("Q2:{:.4f}".format(q2))
print("Q3:{:.4f}".format(q3))
aiq = q3 - q1                       # Amplitude Interquartil = Q3 -Q1
print("AIQ:{:.4f}".format(aiq))
print("q3-q2: {:.4f}".format(q3-q2))
print("q1-q2: {:.4f}".format(abs(q1-q2)))
print("P10:{:.4f}".format(p10))
print("P20:{:.4f}".format(p20))
print("P90:{:.4f}".format(p90))

print()
assim = df_mun_2021['taxa'].skew()
curt  = df_mun_2021['taxa'].kurtosis()
print("Assimetria:{:.2f}".format(assim))
print('Curtose:{:.2f}'.format(curt))
print('media:{:.2f}'.format(media))
print('mediana:{:.2f}'.format(mediana))
print('media-mediana:{:.2f}'.format(media-mediana))
```

## ANÁLISE DOS OUTLIERS

```python
# =====================================
# CÁLCULO DOS LIMITES DOS OUTLIERS
# LIMITE INFERIOR = Q1 - 1,5 X AIQ
# LIMITE SUPERIOR = Q3 + 1,5 X AIQ
# =====================================
```

VARIÁVEL ANO 

```python
lia = q1a - 1.5*(q3a-q1a)
lsa = q3a + 1.5*(q3a-q1a)
```

```python
print("Limite Inferior(lia):{:.4f}".format(lia))
print("Limite Superior(lsa):{:.4f}".format(lsa))
print()
print("OUTLIERS")
for i in media_ano.index:
    if (media_ano.loc[i,] < lia) or (media_ano.loc[i,] > lsa):
     print("ano: {:.4s}".format(i),"media_ano: {:.4f}".format(media_ano.loc[i,]))
```
